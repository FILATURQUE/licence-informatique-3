Partitionez le dataset pour avoir X_train, y_train X_test, y_test

%75 pour l'ensemble d'apprentissage, %25 pour l'ensemble de test

mettez random_state=42 pour avoir tous les mêmes valeurs

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    b_cancer['data'],
    b_cancer['target'],
    test_size = 0.25,
    random_state = 42
)
```


Importez  la méthode du k plus proche voisins



```python
from sklearn.neighbors import KNeighborsClassifier
```

Construisez  le modèle pour le nombre de voisin=1 et pour l'ensemble d'apprentissage



```python
model1 = KNeighborsClassifier(n_neighbors = 1)
```

Affichez le score pour l'ensemble de test

```python
kn1 = model1.fit(X_train, y_train)
score = model1.score(b_cancer['data'],b_cancer['target'])
print("Score:",score)
```

```python
x_new = np.array([1.6e+01, 1.674e+01, 9.873e+01, 6.894e+02, 9.883e-02, 1.364e-01,
                  7.721e-02, 6.142e-02, 1.668e-01, 6.869e-02, 3.720e-01, 8.423e-01,
                  2.304e+00, 3.484e+01, 4.123e-03, 1.819e-02, 1.996e-02, 1.004e-02,
                  1.055e-02, 3.237e-03, 1.676e+01, 2.043e+01, 1.097e+02, 8.569e+02,
                  1.2e-01, 3e-01, 1.856e-01, 1.018e-01, 2.2e-01, 8.8e-02])

x_new = x_new.reshape(1, -1)

print("etiquette :{}".format(kn1.predict(x_new)))
print("target names:",b_cancer['target_names'])
```

Utiliser la méthode cross-validation pour trouver le meilleur modèle (estimateur)

```python
from sklearn.model_selection import cross_val_score

for i in range(1,11):
    scores=cross_val_score(KNeighborsClassifier(i), X_train, y_train, cv=5,scoring="accuracy")
    print("scores:",scores, i)
    
    print("valeur moyenne:", scores.mean())
```

Affichez la matrice de confusion pour ce modèle

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

model_final = KNeighborsClassifier(8)
model_final.fit(X_train, y_train)
y_pred = model_final.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=b_cancer.target_names)
disp.plot(cmap="Blues")
```

```python
from sklearn.preprocessing import MinMaxScaler
model_final = make_pipeline(MinMaxScaler(),KNeighborsClassifier(8))
model_final.fit(X_train, y_train)
y_pred = model_final.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=b_cancer.target_names)
disp.plot(cmap="Blues")
```

```python
model_final = make_pipeline(StandardScaler(),KNeighborsClassifier(8))
model_final.fit(X_train, y_train)
y_pred = model_final.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=b_cancer.target_names)
disp.plot(cmap="Blues")
```

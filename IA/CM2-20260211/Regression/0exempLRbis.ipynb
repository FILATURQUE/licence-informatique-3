```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()
print(model)
```

```python
x = np.array([0,1,2,3,4,5])
y = np.array([2, 3, 4,5,6,7])

print(x)
plt.plot(x, y, marker='o')
```

```python
X = x[:, np.newaxis] # The input data for sklearn is 2D: (samples == 6 x features == 1)



print(X)


```

```python

lr = model.fit(X, y)

print(lr.predict([[10]]))

#faire apprentissage

#aX+b=y

print('lr.coef (a): {}'.format(lr.coef_))
print('lr.intercept (b): {}'.format(lr.intercept_))

print(lr.coef_*10 +lr.intercept_)
```

```python
# Create some simple data
import numpy as np
np.random.seed(0)
X = np.random.random(size=(10, 1))

print(X)


```

```python
print(" x \n {} \n ".format(X[:,0]))

#Etiquette
y = 5 * X[:, 0] + 3 + np.random.normal(size=10) #on ajoute du bruit aléatoire 

print("y \n{} \n".format(y))
```

```python
model = LinearRegression()
lr = model.fit(X, y)

print(' \n lr.coef (a): {}'.format(lr.coef_))
print('\n lr.intercept (b): {}'.format(lr.intercept_))
```

```python

plt.scatter(X[:, 0], y, label='donnee')

yr = np.array(lr.coef_ * X[:, 0] + lr.intercept_)

plt.plot(X[:, 0], yr, '-', c='red', label='lr')
plt.legend()
```

```python

model.score(X,y)
```

```python
model.predict([[1]])


```

```python
# predict y from the data
Xe = np.linspace(0, 6, 10)
Xe = Xe[:, np.newaxis]
ye = model.predict(Xe)


print(" x \n {} \n ".format(Xe[:,0]))



print("y \n{} \n".format(ye))


plt.scatter(Xe[:, 0], ye, label='estimation')

yr = np.array(lr.coef_ * Xe[:, 0] + lr.intercept_)

plt.plot(Xe[:, 0], yr, '-', c='red', label='lr')
plt.legend()
```

**Si on change le modèle**

Régression rigide  (régularisation)  

On impose des contraintes supplémentaires sur les paramètres de régression. Le but de 

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html

```python



from sklearn.linear_model import Ridge

model = Ridge(alpha=1)



lr3 = model.fit(X2,y2)

print(' \n lr.coef (a): {}'.format(lr3.coef_))
print('\n lr.intercept (b): {}'.format(lr3.intercept_))

plt.scatter(X2[:, 0], y2,label='donnees')

yr3 = np.array(lr3.coef_ * X2[:, 0] + lr3.intercept_)

plt.plot(X2[:, 0], yr3, '-', c='red', label='lr')
plt.legend()
```

```python
model.score(X2,y2)
```
